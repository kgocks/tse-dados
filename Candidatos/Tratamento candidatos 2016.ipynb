# Verificar número de linhas duplicadas


```python
import wget
import pandas as pd
import os.path
def get_file(file_name):
    file_path = "input/"+file_name
    if not os.path.exists(file_path):
        url = 'https://s3-sa-east-1.amazonaws.com/datascience2016/data/gzip/candidatos/'+ file_name
        filename = wget.download(url,"input/")
        return pd.read_csv(filename, sep=";",names=cand_h,dtype=str)
    else:
        return pd.read_csv(file_path, sep=";",names=cand_h,dtype=str)

for year in range(2002,2017,2):
    csv = get_file("candidato_"+str(year)+".csv.gz")
    csv_grouped = csv.groupby(['NUM_TURNO', 'NUMERO_CANDIDATO', 'CODIGO_CARGO', 'SIGLA_UE', 
                                    'DESCRICAO_ELEICAO'],as_index=False)["NOME_CANDIDATO"].count()
    print str(year)  + "- Quantidade de candidatos repetidos: "+str(csv_grouped[csv_grouped["NOME_CANDIDATO"]>1].shape[0])


```
