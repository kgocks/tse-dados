```python
import pandas as pd
names = ["DATA_GERACAO",
    "HORA_GERACAO",
    "ANO_ELEICAO",
    "NUM_TURNO",
    "DESCRICAO_ELEICAO",
    "SIGLA_UF",
    "SIGLA_UE",
    "NOME_UE",
    "CODIGO_CARGO",
    "DESCRICAO_CARGO",
    "TIPO_LEGENDA",
    "NUM_PARTIDO",
    "SIGLA_PARTIDO",
    "NOME_PARTIDO",
    "SIGLA_COLIGACAO",
    "NOME_COLIGACAO",
    "COMPOSICAO_COLIGACAO",
    "SEQUENCIA_COLIGACAO"]


### Lendo o CSV
data = pd.read_csv("input/legendas_2010.csv", sep=";", names=names)


data.head()
```

```python
### Agrupando por partido e sequencia coligacao
data_g =data.groupby(["SEQUENCIA_COLIGACAO","SIGLA_PARTIDO","SIGLA_UF","CODIGO_CARGO","DESCRICAO_CARGO"], as_index=False).count()
data_partido = data_g[["SEQUENCIA_COLIGACAO","SIGLA_PARTIDO","SIGLA_UF","CODIGO_CARGO","DESCRICAO_CARGO"]]
data_partido
```

```python
### Retirando os partidos sem coligação
data_partido = data_partido[data_partido["SEQUENCIA_COLIGACAO"]!=-1]

### Agrupando os partidos com a mesma sequencia coligacao
data_partido_g = data_partido.groupby(["SEQUENCIA_COLIGACAO","SIGLA_UF","CODIGO_CARGO"])["SIGLA_PARTIDO"].apply(lambda x: "%s" % ' / '.join(x))
data_partido_g.name = "COMPOSICAO_COLIGACAO2"
```

```python
data_partido_g

```

```python


### Substitiundo os partidos agrupados na tabela original
composicao = data.join(data_partido_g, on=["SEQUENCIA_COLIGACAO","SIGLA_UF","CODIGO_CARGO"]).fillna("#NE#")

composicao
```

```python
### Exportando o csv
composicao["COMPOSICAO_COLIGACAO"] = composicao["COMPOSICAO_COLIGACAO2"]
composicao = composicao.drop(["COMPOSICAO_COLIGACAO2"],1)
```

```python
composicao.to_csv("output/legendas_2010.csv",header=False, index=False,sep=";")
```
