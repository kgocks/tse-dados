## Alguns exemplos de linhas com mais de 5 resultados.

``` mysql
select * from (
    select c.ANO_ELEICAO, c.NUM_TITULO_ELEITORAL_CANDIDATO
    from (
        select count(*) as ocorrencia, NUM_TITULO_ELEITORAL_CANDIDATO,ANO_ELEICAO 
        from detalhe_candidato_2000 
        group by ANO_ELEICAO,NUM_TITULO_ELEITORAL_CANDIDATO
    ) as c where c.ocorrencia>4 AND c.ANO_ELEICAO='2000') as m 
    join detalhe_candidato_2000 as a 
    on a.NUM_TITULO_ELEITORAL_CANDIDATO=m.NUM_TITULO_ELEITORAL_CANDIDATO and 
    a.NUM_TITULO_ELEITORAL_CANDIDATO<>'#NI#' and 
    a.DESC_SIT_TOT_TURNO=""
    order by a.NUM_TITULO_ELEITORAL_CANDIDATO limit 1000; 
```
